```python
moviedf = dataframe.pivot_table(
    index='program_name',
    columns='user_id_maped',
    values='rating').fillna(0)
moviedf.head()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity_matrix = cosine_similarity(moviedf)

similarity_df = pd.DataFrame(
    similarity_matrix,
    index=moviedf.index,
    columns=moviedf.index
)
similarity_df.head()
```

```python
recommendations = similarity_df['Moana'].sort_values(ascending=False)
recommendations = recommendations.drop('Moana')  # drop Moana itself, since it's always perfectly similar to itself
top_5 = recommendations.head(5)

print("Top 5 recommendations for people who watched Moana:")
print(top_5)
```

```python
top_5_pct = (top_5 * 100).round(1)
for name, score in top_5_pct.items():
    print(f"{name}: {score}% match")
```
